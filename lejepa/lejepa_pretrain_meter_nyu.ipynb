This Kaggle notebook configures the run, imports the LeJEPA modules, and starts training.

```python
from __future__ import annotations

import sys
from dataclasses import replace
from pathlib import Path

REPO_ROOT = Path("/kaggle/input/meter-code")
if not REPO_ROOT.exists():
    REPO_ROOT = Path.cwd().parent if Path.cwd().name == "notebooks" else Path.cwd()
if str(REPO_ROOT) not in sys.path:
    sys.path.insert(0, str(REPO_ROOT))

from lejepa_globals import LeJEPAConfig, parse_config
from lejepa_train import run_training
```

```python
config = parse_config()
config = replace(
    config,
    output_dir="/kaggle/working/lejepa",
    local_data_root="nyu_depth_v2",
    model_size="xxs",
    epochs=100,
    batch_size=64,
    global_views=2,
    local_views=2,
    train_limit=None,
)
run_training(config)
```
